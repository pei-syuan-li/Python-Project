## Simulations for Flipping Coins
* Simulating numbers of heads for 100 times with a replication of 200 times
* Plotting the results given the probabilities are 0.3, 0.5 and 0.7

```python
import random
import matplotlib.pyplot as plt
import numpy as np
prob =[0.3, 0.5, 0.7]
colors=['go', 'ro', 'co']
ct = []

def figure(repeat, toss):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.set_xlabel('Numbr of heads', color = 'k' )
    ax1.set_ylabel('Pobability of flipping heads', color = 'k')
    plt.title("Flipping %d coins for %d times" % (toss, repeat))
    for j in range(3):
        hh = []
        # 為增加視覺效果, 增加微小隨機值
        hh = [np.random.binomial(toss, prob[j]) +
              random.random()/4*random.choice([-1,1])
              for r in range(repeat)]
        ct.append(hh)
        plt.plot(hh, np.repeat(prob[j], repeat), colors[j],
                 alpha = 0.7, label = ' p = %.1f' % prob[j])
    plt.xlim(-0.2, toss+1) 
    plt.ylim(prob[0]-0.05, prob[2]+0.05)
    plt.legend(loc = 'upper left', shadow = True)  
    plt.show()
```

* Scatter plot for flipping coins 100 times with a replication of 200 times

```python
figure(repeat = 200, toss = 100)
```

```python
col=['g', 'r', 'c']
def plot_hist(repeat, toss):
    fig2 = plt.figure()
    ax1 = fig2.add_subplot(111)
    ax1.set_xlabel('Numbr of heads', color = 'k' )
    ax1.set_ylabel('Density', color = 'k' )
    plt.xlim(-0.2, toss+1)
    plt.ylim(0, 0.12) 
    plt.title("Histogram of Flipping %d coins" % toss)
    for i in range(len(ct)):
        plt.hist( ct[i], normed= True, color = col[i] ,
                  alpha= 0.7, label =' p = %.1f' % prob[i])
    plt.legend(loc='upper left', shadow=True)
    plt.show()
```

* Plotting the histogram

```python
plot_hist(repeat = 200, toss = 100)
```

* What are the results when repeating 500 times?

```python
ct = []
figure(repeat = 200, toss = 500) 
plot_hist(repeat = 200, toss = 500)
```
